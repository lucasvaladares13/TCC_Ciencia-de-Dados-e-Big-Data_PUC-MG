```python
def generate_image(df, col, path):
    try:
        fig, ax = plt.subplots(1, sharey=True, tight_layout=True)
        df[col].plot(kind='hist',ax=ax, title = col)
        fig.savefig(path+col+'.png', dpi=300)
        plt.close()
    except:
        print(col)
    
```

```python
for col in columns_image:
    generate_image(df,col,path_images)

```
