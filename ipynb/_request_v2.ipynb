```python
def create_partida(file):
    columns = ['data_realizacao','data_realizacao_iso','grupo','hora_realizacao','partida_id','placar','placar_mandante','placar_visitante','rodada','slug','status']

    df = pd.read_json(file)
    df = df[columns].reset_index(drop = True).drop_duplicates()
    file = 'dt_partida_' + str(file.split('.')[0]) + '.xlsx'
    
    df.to_excel(file, index= False)
```

```python
def create_time(file):
    columns = ['partida_id']
    df = pd.read_json(file)
    df = df[columns].reset_index(drop = True).drop_duplicates()
    partida_id = str(df.loc[0,'partida_id'])


    with open(file, encoding='utf-8') as meu_json:
        df_json = json.load(meu_json)
    

    df_mandante = json_normalize(df_json['time_mandante'])
    df_mandante['partida_id'] = partida_id
    df_mandante['flag_time'] = 'Mandante'

    df_visitante = json_normalize(df_json['time_visitante'])
    df_visitante['partida_id'] = partida_id
    df_visitante['flag_time'] = 'Visitante'

    df_time = df_mandante.append(df_visitante)


    file = 'dt_time_' + str(file.split('.')[0]) + '.xlsx'

    df_time.to_excel(file, index= False)



    
```
